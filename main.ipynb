```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

dog = cv2.imread("./images/dog.png")

cat = "CAT"
cat_sum = sum(map(ord, cat))
cat_length = len(cat)

secret = np.zeros((cat_sum, cat_sum, cat_length), np.uint8)

for r in range(cat_sum):
    for c in range(cat_sum):
        secret[r][c] = dog[r*cat_length][c*cat_length]

secret_rgb = cv2.cvtColor(secret, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(secret_rgb)
plt.axis('off')  # Hide the axis
plt.show()
```
